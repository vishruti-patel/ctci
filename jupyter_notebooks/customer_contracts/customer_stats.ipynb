```python
import pandas as pd

df = pd.read_csv('data.csv')
df
    
```

```python
contract= df.groupby('COUNTRY')['CONTRCNT'].sum().rename('contract').reset_index()
contract
```

```python
max_contract = contract['contract'].max()
max_contract
```

```python
max_country = contract.sort_values('COUNTRY', ascending =False).iloc[0]
max_country
```

```python
print(f"Country with the largest number of customers' contracts: \n{max_country}")
```
