```python
import pandas as pd
```

```python
data = {
   'order_id': [101, 102, 103, 104, 105, 106, 107],
   'customer_id': [1, 2, 1, 3, 2, 3, 1],
   'product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Monitor', 'Mouse', 'Laptop'],
   'category': ['Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics'],
   'price': [1200, 25, 300, 45, 280, 25, 1150],
   'quantity': [1, 2, 1, 3, 1, 20, 1],
   'order_date': pd.to_datetime([
       '2024-01-05', '2024-01-07', '2024-01-10',
       '2024-01-12', '2024-01-14', '2024-01-15', '2024-01-18'
   ])
}

# df = pd.read_csv(data.csv)
df = pd.DataFrame(data)
df
```

```python
df['total_rev'] = df['price'] * df['quantity']
df
```

```python
res = df.groupby('customer_id')['total_rev'].sum().rename('revenue').reset_index()
res
```

```python
res_sorted = res.sort_values(by='revenue', ascending = False)
res_sorted
```

```python
import pandas as pd
df = pd.DataFrame(data)

df['revenue'] = df['price'] * df ['quantity']

res = df.groupby('customer_id')['revenue'].sum().rename('total_revenue').reset_index()
res_sorted = res.sort_values(by = 'total_revenue', ascending = False)
res_sorted
```
