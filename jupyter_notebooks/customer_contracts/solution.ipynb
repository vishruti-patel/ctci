# Total Revenue Problem

```python
data = {
   'order_id': [101, 102, 103, 104, 105, 106, 107],
   'customer_id': [1, 2, 1, 3, 2, 3, 1],
   'product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Monitor', 'Mouse', 'Laptop'],
   'category': ['Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics'],
   'price': [1200, 25, 300, 45, 280, 25, 1150],
   'quantity': [1, 2, 1, 3, 1, 2, 1],
   'order_date': pd.to_datetime([
       '2024-01-05', '2024-01-07', '2024-01-10',
       '2024-01-12', '2024-01-14', '2024-01-15', '2024-01-18'
   ])
}


df = pd.DataFrame(data)
df
```

```python
df['revenue'] = df['price'] * df['quantity']
df
```

```python
df2 = df.groupby('customer_id')['revenue'].sum() \
    .rename('total_revenue') \
    .reset_index()
df2
```

```python
# df2['total_revenue'] = df2['total_revenue'] > 1000
df2 = df2[df2['total_revenue'] > 1000]
df2
```

```python
df2.sort_values(by='total_revenue', ascending=False)
df2
```
